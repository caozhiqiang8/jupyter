# 初始化数据库

```python
task_type = [1,2,3,4,6,7,10,13,14,15]
type_name = ['学资源','讨论','单题','测验','微课程','一般任务','直播课','答题卡','个性化','先声']
task_df = []
for i in task_type:
    if i ==7:
        sql = '''
            SELECT DATE_FORMAT(tt.c_time,'%Y-%m-%d') as time,count(tt.task_id) as '{}'
            FROM tp_task_info tt ,tp_course_info tc,franchised_school_info fr
            WHERE tc.DC_SCHOOL_ID = fr.school_id  and tc.course_id = tt.COURSE_ID  and fr.school_id >50000 and fr.enable = 0 and tc.LOCAL_STATUS = 1 
            and fr.school_type in (3,4) and tt.task_type in (7,8,9)
            GROUP BY time
            ORDER BY time DESC
            '''.format(i)
        task = mysqlDB(sql)
        task_df.append(task)
    elif i == 6:
        sql = '''
            SELECT DATE_FORMAT(tt.c_time,'%Y-%m-%d') as time,count(tt.task_id) as '{}'
            FROM tp_task_info tt ,tp_course_info tc,franchised_school_info fr
            WHERE tc.DC_SCHOOL_ID = fr.school_id  and tc.course_id = tt.COURSE_ID  and fr.school_id >50000 and fr.enable = 0 and tc.LOCAL_STATUS = 1 
            and fr.school_type in (3,4) and ( ( tt.task_type =13 and video_id is not  null) or tt.task_type =6)
            GROUP BY time
            ORDER BY time DESC
            '''.format(i)
        task = mysqlDB(sql)
        task_df.append(task)
    elif i ==13:
        sql = '''
            SELECT DATE_FORMAT(tt.c_time,'%Y-%m-%d') as time,count(tt.task_id) as '{}'
            FROM tp_task_info tt ,tp_course_info tc,franchised_school_info fr
            WHERE tc.DC_SCHOOL_ID = fr.school_id  and tc.course_id = tt.COURSE_ID  and fr.school_id >50000 and fr.enable = 0  and tc.LOCAL_STATUS = 1 
            and fr.school_type in (3,4) and tt.task_type =13 and video_id is null
            GROUP BY time
            ORDER BY time DESC
            '''.format(i)
        task = mysqlDB(sql)
        task_df.append(task)
    else:
        sql = '''
            SELECT DATE_FORMAT(tt.c_time,'%Y-%m-%d') as time,count(tt.task_id) as '{}'
            FROM tp_task_info tt ,tp_course_info tc,franchised_school_info fr
            WHERE tc.DC_SCHOOL_ID = fr.school_id  and tc.course_id = tt.COURSE_ID  and fr.school_id >50000 and fr.enable = 0 and tc.LOCAL_STATUS = 1 
            and fr.school_type in (3,4) and tt.task_type = {}
            GROUP BY time
            ORDER BY time DESC
            
            '''.format(i,i)
        task = mysqlDB(sql)
        task_df.append(task)
every_task = reduce(lambda left, right: pd.merge(left, right, on=['time'], how='outer'), task_df)
  
```

```python
every_task.columns=type_name = ['date','xzy','tl','dt','cy','wkc','yb','zbk','dtk','gxh','xs',]
every_task = every_task.fillna(0)
every_task.to_sql(name='every_day_task',con=con, if_exists='replace', index=False)
```
