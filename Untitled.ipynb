```python
import time
import  json
import  pandas as pd
from elasticsearch import Elasticsearch
```

```python
def es_connect(index, body):
    es_hosts = str("69.230.239.155,43.192.117.34").split(",")
    es = Elasticsearch(es_hosts)
    res = es.search(index=index, body=body)
    res = json.loads(json.dumps(res))
    return res
```

```python
hits_list = []
b_num = 0
e_num = 10000
while True:
    body = '''
    {
      "query": {
        "bool": {
          "must": [
            {
              "term": {
                "url.keyword": {
                  "value": "http://school-api.ai-classes.com/app-common-service/checkAppVersionWithCache.do"
                }
              }
            }
          ]
        }
      },
      "from":%s,
      "size": %s
    }
    '''%(b_num,e_num)
    index = 'action_logs_20230305'

    data = es_connect(index=index, body=body)
    total = data['hits']['total']
    print()
    if total < b_num:
        break
    else:
        hits = data['hits']['hits']
        hits_list.extend(hits)

    b_num+=10000
    e_num+=10000
    print(b_num)
    time.sleep(1)
hits_list
```

```python
df_list = []

for i in hits_list:
    sourse_data = i['_source']
    sourse_data_df = pd.json_normalize(sourse_data)

    param_json = json.loads(sourse_data['param_json'])
    param_json_df = pd.json_normalize(param_json)
    df = pd.merge(sourse_data_df,param_json_df,how='left')
    df_list.append(df)

df = pd.concat(df_list,axis=0,ignore_index=True)
print(df)
```
